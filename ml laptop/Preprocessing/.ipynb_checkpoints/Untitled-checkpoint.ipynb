```python
df['Marka'] = data['brand']
df['Ekran Boyutu'] = data['Ekran Boyutu']
df['Çözünürlük'] = data['Çözünürlük']
df['Ekran Yenileme Hızı'] = data['Ekran Yenileme Hızı']
df['İşlemci Tipi'] = data['İşlemci Tipi']
df['İşlemci Çekirdek Sayısı'] = data['İşlemci Çekirdek Sayısı']
df['İşlemci Modeli'] = data['İşlemci Modeli']
df['Ram (Sistem Belleği)'] = data['Ram (Sistem Belleği)']
df['SSD Kapasitesi'] = data['SSD Kapasitesi']
df['Ekran Kartı'] = data['Ekran Kartı']
df['Ekran Kartı Hafızası'] = data['Ekran Kartı Hafızası']
df['İşletim Sistemi'] = data['İşletim Sistemi']
df['Cihaz Ağırlığı'] = data['Cihaz Ağırlığı']
df['Fiyat'] = data['price']
```

```python
df.head(10)
```

```python
df['Fiyat'] = df['Fiyat'].str.replace(r',.*', '', regex=True) # Virgül sonrasındaki sayıları siler
df['Fiyat'] = df['Fiyat'].str.replace('.', '', regex=False) # Noktayı siler
df['Fiyat'] = df['Fiyat'].str.replace('TL', '', regex=False) # "TL" ifadesini siler
df["Fiyat"] = df.Fiyat.astype(float) # "Fiyat" sütununu float türüne dönüştürür
```
